```python
import pandas as pd
```

```python
from google.colab import drive
drive.mount('/drive')
```

```python
path = "/drive/My Drive/Data-driven/Insper/Siscan_Mamografia_por_pacientes.xlsx"
testes_mamo = pd.read_excel(path, sheet_name = 'UFs e Brasil')
testes_mamo
```

```python
# Maneira 1 
```

```python
df_tidy_mamo = testes_mamo.melt('Ano', var_name='localizacao', value_name='numero_testes')
```

```python
df_tidy_mamo
```

```python
# Maneira 2
```

```python
df_tidy_mamo2 = testes_mamo.set_index('Ano').rename_axis(columns='localizacao').unstack() \
        .rename('numero_testes').reset_index()
```

```python
df_tidy_mamo2
```

```python
df_tidy_mamo.to_excel('/drive/My Drive/Data-driven/Insper/tidy_siscan_mamografia_por_pacientes_ufs_brasil.xlsx',sheet_name='Sheet1',index=False)
df_tidy_mamo.to_csv('/drive/My Drive/Data-driven/Insper/tidy_siscan_mamografia_por_pacientes_ufs_brasil.csv',index=False)
```
